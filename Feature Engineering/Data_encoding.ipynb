1. Nomimal/OHE Encoding

One hot encoding, also known as nomial encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical varibale 'color' with three possible values(red,green,blue), we canrepresent it using one hot encoding as follows:

1. Red:[1,0,0]
2. Green:[0,1,0]
3. Blue:[0,0,1]

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder

df=pd.DataFrame({'color': ['red','blue','green','green','blue','red']})
```

```python
df.head()
```

```python
#Create an instance of onehotencoder

encoder=OneHotEncoder()
```

```python
##perform fit and transform

encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
# for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```
